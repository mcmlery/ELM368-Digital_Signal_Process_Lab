### SORU-3

# I)

```python
#grafiklerin çıktısı alınırken real ve imaginer kısımlarının çıktıları ayrı ayrı alındı
x2 = np.exp(1j*np.pi/5)
n=np.arange(0,15)
x2_i=np.zeros(15) # sabit x2 değerini bir diziye atamak için 15 elemanlı 0 dizisi oluşturuldu
for i in range(0,15): #oluşturulan diziye atamalar yapıldı
    x2_i[i] = x2.imag
plt.title('x2[n] imag') #grafiğin başlığı belirtildi
plt.stem(n,x2_i) # grafik çıktısı alındı
plt.show()
```

```python
x2_r=np.zeros(15)
for i in range(0,15): #aynı işlem tekrarlandı
    x2_r[i] = x2.real
plt.title('x2[n] real') 
plt.stem(n,x2_r)
plt.show()
```

# II)

```python
n=np.arange(-10,4) #ilk kısımdaki işlemlerin hepsi burası için de gerçekleştirildi
x6 = np.cos(np.pi*n/5)+1j*np.sin(np.pi*n/5)
plt.title('x6[n] imag')
plt.stem(n,x6.imag)
plt.show()
```

```python
n=np.arange(-10,4)
x6 = np.cos(np.pi*n/5)+1j*np.sin(np.pi*n/5)
plt.title('x6[n] real')
plt.stem(n,x6.real)
plt.show()
```
