```python
from scipy.io.wavfile import read # Ses dosyasının okunması için gerekli kütüphane
Fs, x = read("ELM368 Ödev-6 odev6.wav")

import matplotlib.pyplot as plt
import numpy as np 
```

```python
xn = x[:,1] 
N = len(x)
n=np.arange(0,N)
plt.figure()
plt.stem(n,xn) 
plt.xlabel("n (örnek)") 
plt.ylabel("x[n]")
M = 2
xn_d = xn[np.arange(0, np.size(xn, 0), M)] 
N_d = len(xn_d)
n_d = np.arange(0,N_d) 
plt.figure()
plt.stem(n_d, xn_d) 
plt.ylabel('$x_d[n]$') 
plt.xlabel('n (örnek)') 
plt.show()
```

```python
from IPython.lib.display import Audio
print('x[n] işareti için:')
display(Audio(xn, rate=Fs))
print('x_d[n] işareti için:')
display(Audio(xn_d, rate=Fs/M))
```

```python
L = 2
N = len(xn_d)
N_u = N*L
xn_u = np.zeros(N_u) 
xn_u[np.arange(0,len(xn_u),L)] = xn_d
n_u = np.arange(0,N_d) 
hn = np.array([0,1/2,1,1,1/2,0])
xn_u = np.convolve(xn_u,hn,'full') 
n_u = np.arange(0, len(xn_u))
plt.figure()
plt.stem(n_u,xn_u)
plt.xlabel("n (örnek)") 
plt.ylabel("$x_u[n]$")
```

```python
print('x_u[n] işareti için:')
display(Audio(xn_u, rate=Fs))
```

### Ses ilk olarak seyrek örneklendi ardından seyrek örneklediğimiz sesi tekrar sık örnekleyerek ilk haline geri döndürdük
### M=2 ve L=2 ile up and down sampling yaptığımız için kalite kaybını farkedemeyiz (örneklerken fazla etkisi olmayan örnekleme gerçekleştirdik)
